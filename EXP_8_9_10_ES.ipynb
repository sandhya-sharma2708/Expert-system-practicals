```python
# Practical: Backward Chaining in Expert System

# --- Knowledge Base ---
rules = {
    "Wet": ["Raining"],           # If Raining then Wet
    "Slippery": ["Wet"],          # If Wet then Slippery
    "PlantsGrow": ["Raining"],    # If Raining then PlantsGrow
}

facts = ["Raining"]  # Known facts

# --- Backward Chaining Function ---
def backward_chaining(goal):
    print(f"Trying to prove: {goal}")

    # If goal is a known fact
    if goal in facts:
        print(f"✔ {goal} is a known fact.")
        return True

    # If goal can be concluded by a rule
    if goal in rules:
        conditions = rules[goal]
        print(f"Rule found: {goal} <- {conditions}")
        for condition in conditions:
            if not backward_chaining(condition):
                print(f"✖ Failed to prove {goal} because {condition} is not true.")
                return False
        print(f"✔ {goal} has been proven true.")
        facts.append(goal)
        return True

    # If no rule or fact supports the goal
    print(f"✖ No rule or fact found for {goal}.")
    return False


# --- Main Program ---
query = "Slippery"

print("\n--- BACKWARD CHAINING EXPERT SYSTEM ---\n")
result = backward_chaining(query)

print("\n--- RESULT ---")
if result:
    print(f"The query '{query}' is TRUE according to the knowledge base.")
else:
    print(f"The query '{query}' cannot be proven true.")
```

```python
  # Forward Chaining Algorithm in Expert System
# Rule base (each rule: {'if': [conditions], 'then': 'conclusion'})
rules = [
    {'if': ['A'], 'then': 'B'},
    {'if': ['B', 'C'], 'then': 'D'},
    {'if': ['D'], 'then': 'E'},
    {'if': ['A', 'F'], 'then': 'C'}
]
# Initial facts
facts = ['A', 'F']
# Forward Chaining Function
def forward_chaining(rules, facts):
    derived = set(facts)
    new_fact_added = True
    while new_fact_added:
        new_fact_added = False
        for rule in rules:
            if all(cond in derived for cond in rule['if']) and rule['then'] not in derived:
                print(f"Rule fired: IF {rule['if']} THEN {rule['then']}")
                derived.add(rule['then'])
                new_fact_added = True
    return derived

# Run the forward chaining process
result = forward_chaining(rules, facts)

print("\nFinal Derived Facts:")
print(result)
```

```python
# Naive Bayes Model implementation without using sklearn
# Example: Weather prediction expert system

# Training data
# Outlook: Sunny=0, Overcast=1, Rainy=2
# Temperature: Hot=0, Mild=1, Cool=2
# Humidity: High=0, Normal=1
# Windy: False=0, True=1
dataset = [
    [0, 0, 0, 0, 0],  # Sunny, Hot, High, False, No
    [0, 0, 0, 1, 0],  # Sunny, Hot, High, True, No
    [1, 0, 0, 0, 1],  # Overcast, Hot, High, False, Yes
    [2, 1, 0, 0, 1],  # Rainy, Mild, High, False, Yes
    [2, 2, 1, 0, 1],  # Rainy, Cool, Normal, False, Yes
    [2, 2, 1, 1, 0],  # Rainy, Cool, Normal, True, No
    [1, 2, 1, 1, 1],  # Overcast, Cool, Normal, True, Yes
    [0, 1, 0, 0, 0],  # Sunny, Mild, High, False, No
    [0, 2, 1, 0, 1],  # Sunny, Cool, Normal, False, Yes
    [2, 1, 1, 0, 1],  # Rainy, Mild, Normal, False, Yes
]

# Split features and labels
X = [row[:-1] for row in dataset]
y = [row[-1] for row in dataset]

# Calculate prior probabilities
total = len(y)
yes_prob = y.count(1) / total
no_prob = y.count(0) / total

# Function to calculate conditional probability
def cond_prob(attr_index, attr_value, label):
    count_label = 0
    count_attr = 0
    for i in range(total):
        if y[i] == label:
            count_label += 1
            if X[i][attr_index] == attr_value:
                count_attr += 1
    return count_attr / count_label if count_label else 0

# Test data (e.g., Sunny, Cool, High, True)
test = [0, 2, 0, 1]

# Compute probabilities for Yes and No
prob_yes = yes_prob
prob_no = no_prob

for i in range(len(test)):
    prob_yes *= cond_prob(i, test[i], 1)
    prob_no *= cond_prob(i, test[i], 0)

# Normalize and decide
if prob_yes > prob_no:
    print("Predicted Decision: PLAY")
else:
    print("Predicted Decision: DON'T PLAY")

```
